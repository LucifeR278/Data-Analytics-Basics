### Python loc() fucntion

1. Label based data selecting method(pass the name of the row or column which we want to select.)

2. includes the last element of the range passed it.

3. accept boolean data 

#### Example 1:Selecting Data According to Some Conditions



```python
# selecting cars with brand 'Maruti' and Milage>25
data.loc[(data.Brand=='Maruti') & (data.Mileage>25)]
```

#### Example 2: Selecting a Range of Rows From the DataFrame

```python
data.loc[2:5]
```

##### Example 3: Updating the Value of Any Column

```python
# updating values of Mileage if Year<2015
data.loc[(data.Year<2015),['Mileage']]=22
```

```python
data.loc[(data.Year<2015)]
```

### Python iloc() function

1. indexed-based selecting method

2. does not include the last element of the range passed it in it.

3. does not accept boolean data 

```python
data.iloc[[2,5,3,4]]
```

```python
# selecting rows from 1 to 4 and columns from 2 to 4
data.iloc[1:5,2:5]
```

## Pandas dataframe.groupby() Method

Pandas groupby is used for grouping the data according to the categories and applying a function to the categories. It also helps to aggregate data efficiently

Example 1: Use Pandas groupby function to group the data based on the “Team”. 

```python
import pandas as pd
df=pd.read_csv('nba.csv')
```

```python
df.head()
```

```python
## applying groupby() function to 
## group the data on team value.
gk=df.groupby('Team')

# Let's print the first entries
# in all the groups formed
```

```python
gk.first()
```

```python
# Finding the values contained in the group
gk.get_group('Boston Celtics')
```

```python
# group baseds on more than one category

gkk=df.groupby(['Team','Position'])

# Print the first Value in each group
gkk.first()
```

After performing a groupby() operation in Pandas, you typically apply an aggregation function (like sum(), mean(), etc.) to get a new DataFrame with aggregated results.

```python
import pandas as pd
data={
    'Category':['A','B','A','B','A'],
    'Value':[10,20,15,25,30]

}
df=pd.DataFrame(data)
```

```python
df.groupby('Category').sum().reset_index()
```

```python
data = {
    'Category': ['A', 'B', 'A', 'B', 'A'],
    'Subcategory': ['X', 'Y', 'X', 'Y', 'X'],
    'Value': [10, 20, 15, 25, 30]
}

df=pd.DataFrame(data)
grouped = df.groupby(['Category', 'Subcategory']).mean().reset_index()
print("grouap")
```

```python
import sys
import csv
import operator
def sort_file(read_filename,write_filename,column):
    data=csv.reader(open(read_filename),delimiter=',')
    sortedlist=sorted(data,key=operator.itemgetter(0))
    filewriter=csv.writer(open(write_filename,"w"),delimiter=',')
    for row in sortedlist:
        filewriter.writerow(row)
```
